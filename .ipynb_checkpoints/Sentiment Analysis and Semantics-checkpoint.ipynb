```python
import spacy
import nltk
```

```python
nlp=spacy.load('en_core_web_md')
```

```python
nlp(u'Nepal').vector
```

```python
nlp('Nepal is a beautiful country filled with mountains and valleys').vector
```

```python
# to see similarities between two words or phrases we can use 
doc=nlp('mountains snow Ocean')
```

```python
for tokens in doc:
    for token1 in doc:
        print(tokens, token1, tokens.similarity(token1))
```

```python
doc=nlp('I like apples. The taste sweet.Marcially')
```

```python
for tokens in doc.sents:
    for token1 in doc.sents:
        print(tokens, token1, tokens.similarity(token1))
```

```python
# to find the l2 norm of the vector 
for tokens in doc:
    print (tokens, tokens.vector_norm)
```

```python
#to find if the token has a vector
for tokens in doc:
    print (tokens, tokens.vector_norm, tokens.has_vector)
```

# Vader sentiment Analysis

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
a='Today is a sunny and cold day'
```

```python
sid.polarity_scores(a)
```

```python
b='You are a rat'
```

```python
sid.polarity_scores(b)
```

```python
c='The actors were really good and they did their job really well. However, the direction of the movie was bad which ruined the overall movie.'
```

```python
sid.polarity_scores(c)
```
